```python
# degree_income = []
# degree_outcome = []

# def membership_income():
    
#     for x in np.arange(0, 25, 0.1):
    
#         if x <= 4:
#             degree_income.append(("less", 1)) 
#         if 4 < x <= 5:
#             degree_income.append(("less", -(x - 5)/(5 - 4))) 
#         if 4 < x < 5:
#             degree_income.append(("enough", (x - 4)/(5 - 4))) 
#         if 5 <= x <= 7:
#             degree_income.append(("enough", 1)) 
#         if 7 < x <= 8:
#             degree_income.append(("enough", -(x - 8)/(8 - 7))) 
#         if 7 < x < 8:
#             degree_income.append(("more", (x - 7)/(8 - 7))) 
#         if x >= 8:
#             degree_income.append(("more", 1)) 
    
#     return degree_income
    
# def membership_outcome():
    
#     # Less outcome
#     # Enough outcome
#     # More outcome
#     for x in np.arange(0, 25, 0.1):
    
#         if x <= 2.5:
#             degree_outcome.append(("less", 1)) 
#         if 2.5 < x <= 5:
#             degree_outcome.append(("less", -(x - 5)/(5 - 2.5))) 
#         if 2.5 < x < 5:
#             degree_outcome.append(("enough", (x - 2.5)/(5 - 2.5))) 
#         if 5 <= x <= 7.5:
#             degree_outcome.append(("enough", 1)) 
#         if 7.5 < x <= 9:
#             degree_outcome.append(("enough", -(x - 9)/(9 - 7.5))) 
#         if 7.5 < x < 9:
#             degree_outcome.append(("more", (x - 7.5)/(9 - 7.5))) 
#         if x >= 9:
#             degree_outcome.append(("more", 1))
        
#     return degree_outcome
    
def fuzz_income(x):
    
    fuzzy_sets_income = {'less': 0, 'enough': 0, 'more': 0}

    # Flat line
    if x <= 4:
        fuzzy_sets_income['less'] = 1
    if 5 <= x <= 7:
        fuzzy_sets_income['enough'] = 1
    if x >= 8:
        fuzzy_sets_income['more'] = 1
            
    # Slope
    if 4 < x <= 5:
        fuzzy_sets_income['less'] = -(x - 5)/(5 - 4)
    if 4 < x < 5:
        fuzzy_sets_income['enough'] = (x - 4)/(5 - 4)
    if 7 < x <= 8:
        fuzzy_sets_income['enough'] = -(x - 8)/(8 - 7)
    if 7 < x < 8:
        fuzzy_sets_income['more'] = (x - 7)/(8 - 7)
        
    return fuzzy_sets_income
        
        
def fuzz_outcome(x): 
    
    fuzzy_sets_outcome = {'less': 0, 'enough': 0, 'more': 0}
    
    # Flat line
    if x <= 2.5:
        fuzzy_sets_outcome['less'] = 1
    if 5 <= x <= 7.5:
        fuzzy_sets_outcome['enough'] = 1
    if x >= 9:
        fuzzy_sets_outcome['more'] = 1
        
    # Slope
    if 2.5 < x <= 5:
        fuzzy_sets_outcome['less'] = -(x - 5)/(5 - 2.5)
    if 2.5 < x < 5:
        fuzzy_sets_outcome['enough'] = (x - 2.5)/(5 - 2.5)
    if 7.5 < x <= 9:
        fuzzy_sets_outcome['enough'] = -(x - 9)/(9 - 7.5)
    if 7.5 < x < 9: 
        fuzzy_sets_outcome['more'] = (x - 7.5)/(9 - 7.5)
        
    return fuzzy_sets_outcome
```

```python
# Membership degree lineplot
fig, axs = plt.subplots(2, figsize=(10, 7), sharey=True)
fig.suptitle('Membership Function', fontsize='large', fontweight='bold')
axs[0].plot(range(20), [fuzz_income(x)['less'] for x in range(20)], 'r')
axs[0].plot(range(20), [fuzz_income(x)['enough'] for x in range(20)], 'g')
axs[0].plot(range(20), [fuzz_income(x)['more'] for x in range(20)], 'b')

axs[0].plot(0, 0, 'r', linewidth=1.5, label='less')
axs[0].plot(0, 0, 'g', linewidth=1.5, label='enough')
axs[0].plot(0, 0, 'b', linewidth=1.5, label='more')
axs[0].legend()

axs[1].plot(range(20), [fuzz_outcome(x)['less'] for x in range(20)], 'r')
axs[1].plot(range(20), [fuzz_outcome(x)['enough'] for x in range(20)], 'g')
axs[1].plot(range(20), [fuzz_outcome(x)['more'] for x in range(20)], 'b')

axs[1].plot(0, 0, 'r', linewidth=1.5, label='less')
axs[1].plot(0, 0, 'g', linewidth=1.5, label='enough')
axs[1].plot(0, 0, 'b', linewidth=1.5, label='more')
axs[1].legend()

axs[0].set_ylabel('Income membership degree')
axs[1].set_xlabel('x value')
axs[1].set_ylabel('Outcome membership degree')
```

```python
# Fuzzify dataset
fuzzified_data = []

for i in range(len(df)):
    fuzzed = {'income': 0, 'outcome': 0}
    
    fuzzed['income'] = fuzz_income(df['Penghasilan'][i])
    fuzzed['outcome'] = fuzz_outcome(df['Pengeluaran'][i])
    
    fuzzified_data.append(fuzzed)
```

## Inferences

Here are the rules I've made:

```python
# def rules(income, outcome):
#     if income == "less" and outcome == "less":
#         return "accepted"
#     elif income == "less" and outcome == "enough":
#         return "accepted"
#     elif income == "less" and outcome == "more":
#         return "accepted"
#     elif income == "enough" and outcome == "less":
#         return "rejected"
#     elif income == "enough" and outcome == "enough":
#         return "consideration"
#     elif income == "enough" and outcome == "more":
#         return "accepted"
#     elif income == "more" and outcome == "less":
#         return "rejected"
#     elif income == "more" and outcome == "enough":
#         return "rejected"
#     elif income == "more" and outcome == "more":
#         return "consideration"
    
rules = {
    ('less', 'less'): 'accepted',
    ('less', 'enough'): 'accepted',
    ('less', 'more'): 'accepted',
    ('enough', 'less'): 'rejected',
    ('enough', 'enough'): 'consideration',
    ('enough', 'more'): 'accepted',
    ('more', 'less'): 'rejected',
    ('more', 'enough'): 'rejected',
    ('more', 'more'): 'consideration'
}
```

```python
def inference(fuzzed_data):
    result = {'accepted': 0, "rejected": 0, "consideration": 0}
    
    for income in fuzzed_data['income'].keys():
        for outcome in fuzzed_data['outcome'].keys():
            min_vals = min(fuzzed_data['income'][income], fuzzed_data['outcome'][outcome])
            
            output = rules[(income, outcome)]
            
            if min_vals > result[output]:
                result[output] = min_vals
                
#     for key, value in result.items():
#         print(key, value)

    return result
```

```python
inferenced = []

for fuzzed in fuzzified_data:
    inferenced.append(inference(fuzzed))
```

## Defuzzification
Sugeno inference system

```python
deffuzy = {'accepted': 100, 'rejected': 0, 'consideration': 50}

def defuzzification(inferences, deffuzy):
    numerator, denominator = 0, 0
    
    for output in deffuzy.keys():
        numerator += (inferences[output]) * deffuzy[output]
        denominator += inferences[output]
        
    return numerator/denominator
```

```python
final = []

for i in inferenced:
    final.append(defuzzification(i, deffuzy))
```
